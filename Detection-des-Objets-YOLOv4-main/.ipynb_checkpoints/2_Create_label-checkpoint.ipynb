### Some functions definitions

```python
# this function contains the value (x,y,width,height) for each bounding boxes
def read_region_shape_attributes(attributes):
    if isinstance(attributes, str):
        attributes = eval(attributes)
    
    xmin = attributes["x"]
    ymin = attributes["y"]
    
    width = attributes["width"]
    height = attributes["height"]
    
    xcenter = xmin + width/2
    ycenter = ymin + height/2
    
    return xcenter, ycenter, width, height
```

```python
# this function help us to get the heights and weights of an given image
def get_image_width_height(image_path):
    image = cv2.imread(image_path)
    image_shape = image.shape
    image_height, image_width = image_shape[0], image_shape[1]
    
    return image_width, image_height
```

```python
# here we gonna read the class attriibuted to each bounding boxes
def read_region_attributes(attributes):
    try:
        if isinstance(attributes, str):
            attributes = eval(attributes)

        class_name = attributes["class"]
# if there is no class_name we gonna attribute a -1, that we should clean after, 
# the main goal is all the bounding boxes should have a class
    except:
        class_name = -1
    
    return class_name
```

### Check existing classes

```python
df_labels["class_name"] = df_labels.region_attributes.apply(read_region_attributes)
```

```python
df_labels.class_name.unique()
```

```python
# here we select all the ligne that has (-1) in classe name
df_labels[df_labels.class_name == -1]
```

There were some images with rectangles without classes. We could go back and label them. But for now we will just skip them.

```python
# here we regroup all bad filenames, the class that has -1 in calsss_name in a unique dataframe
bad_filenames = df_labels[df_labels.class_name == -1].filename.unique()
```

```python
# and now we're gonig to clean the data, we taking all the data except the ones that has -1 in class_name
df_labels_clean = df_labels[~df_labels.filename.isin(bad_filenames)]
```

```python
# and as we see that the data has been cleaned
len(df_labels_clean.filename.unique())
```

### Class mapping

```python
# here we define the diffrent class that we have in our project, diffrent names but same id to make it unique
class_mapping = {
    "velo":0,
    'vÃ©lo':0,
    "piéton":1,
    "piÃ©ton":1,
    "pieton":1,
    "trottinette":2, 
    "masque":3, 
    "mal_mis":4, 
    "pas_de_masque":5,
    "pas_de_masque ":5
}
```

### Prepare labels

```python
from shutil import copy
import os
```

```python
all_filenames = df_labels_clean.filename.unique()
```

```python
source_image_folder = "/content/drive/MyDrive/YOLO_train/student_5"
```

```python
# define the ouptub folder where we going to store the 
output_folder = "/content/drive/MyDrive/YOLO_train/images"
```

```python
for filename in all_filenames:
    df_slice = df_labels_clean[df_labels_clean.filename == filename]
    img_path = os.path.join(source_image_folder, filename)
    
    copy(img_path, output_folder)
    
    img_width, img_height = get_image_width_height(img_path)
    
    label_name = '.'.join(filename.split('.')[:-1]) + ".txt"
    label_path = os.path.join(output_folder, label_name)
    

    
    label_str = ""
    if not df_slice.iloc[0].region_count == 0:
        for _, row in df_slice.iterrows():
            class_name = row["class_name"]
            class_id = class_mapping[class_name]
            
            xcenter, ycenter, width, height = read_region_shape_attributes(row["region_shape_attributes"])
            
            xcenter_norm = float(xcenter) / img_width
            ycenter_norm = float(ycenter) / img_height
            
            width_norm = float(width) / img_width
            height_norm = float(height) / img_height
            
            bbox_line = "{} {:.3f} {:.3f} {:.3f} {:.3f}".format(class_id, xcenter_norm, ycenter_norm, width_norm, height_norm)
            label_str += bbox_line
            label_str += "\n"
        
    with open(label_path, "w") as output:
        output.write(label_str)
```
